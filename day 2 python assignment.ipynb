# ques 1 -- 5 different functions in string

```python
str="mahalakshmi learning python"
print(str)
print(str.upper())
print(str.split())
print(str.capitalize())
print(str.replace('a','&'))
```

# ques 2 -- 5 diff functions in list


```python
list=[1,3,5.6,8,"maha","a",["a",2,"n"]]
print(list)
list.append(9)
print(list)
print(list[2:6])
list.remove(3)
print(list)
list.extend([1,2])
print(list)
list.insert(1,"m")
print(list)
```
